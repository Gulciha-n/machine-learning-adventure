```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings("ignore", category=DeprecationWarning)
warnings.filterwarnings("ignore", category=FutureWarning)
```

```python
df = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/ml/USArrests.csv', index_col=0)
```

**linkage: Hiyerarşik kümeleme yapısını oluşturur (veri noktalarının hangi sırayla birleştirileceğini belirler).**

**dendrogram: linkage çıktısına göre bir ağaç diyagramı (dendrogram) çizer, böylece kümeleme süreci ve kümeler arası ilişki görselleştirilir.**


```python
from scipy.cluster.hierarchy import linkage
from scipy.cluster.hierarchy import dendrogram
```

```python
hc_complete = linkage(df, 'complete')
hc_average = linkage(df, 'average')
```

# **Complete Method**

```python
plt.figure(figsize=(10,5))
plt.title('US Arrests Dendograms')
plt.xlabel('Observations')
plt.ylabel('Distance')
dendrogram(hc_complete, labels=df.index, leaf_rotation=90)
plt.show()
```

```python
plt.figure(figsize=(10,5))
plt.title('US Arrests Dendograms')
plt.xlabel('Observations')
plt.ylabel('Distance')
dendrogram(hc_complete,
           truncate_mode="lastp",
           p = 4,                  # küme sayısı
           show_contracted= True,  # kaçar eleman
           )
plt.show()
```

```python
plt.figure(figsize=(10,5))
plt.title('US Arrests Dendograms')
plt.xlabel('Observations')
plt.ylabel('Distance')
dendrogram(hc_complete,
           truncate_mode="lastp",
           p = 15,                  # küme sayısı
           show_contracted= True,   # kaçar eleman
            )
plt.show()
```

# **Average Method**

```python
plt.figure(figsize=(10,5))
plt.title('US Arrests Dendograms')
plt.xlabel('Observations')
plt.ylabel('Distance')
dendrogram(hc_average,
           labels=df.index,
           leaf_rotation=90)
plt.show()
```

```python
plt.figure(figsize=(10,5))
plt.title('US Arrests Dendograms')
plt.xlabel('Observations')
plt.ylabel('Distance')
dendrogram(hc_average,
           truncate_mode="lastp",
           p = 15,                  # küme sayısı
           show_contracted= True,  # kaçar eleman
           )
plt.show()
```
